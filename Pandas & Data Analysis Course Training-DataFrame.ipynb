### .apply method on DataFrame

```python
#Column wise

df_n['Age'] = df_n['Age'].apply(lambda x: x*2 if x > 50 else x)

df_n
```

```python
#Row Wise

#Helps in matching conditions based on multiple columns of a row

def age_check(row):
    age=row[3]
    fare=row[5]
    
    if age > 50 and fare < 20:
        return 'Yes'
    else:
        return 'No'

df_n.apply(age_check, axis=1)
```
