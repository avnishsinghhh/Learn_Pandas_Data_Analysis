### Using pivot_table() Method to create aggregate summary of DataFrame

```python
df_3 = pd.read_csv('test.csv')

df_3
```

```python
df_3.pivot_table(values='Age', index='Sex', aggfunc = 'count')
```

```python
df_3.pivot_table(values='Fare', index='Sex', aggfunc = 'sum')
```

```python
#We can group by data by specifying column names in index parameter
df_3.pivot_table(values='Age', index=['Pclass','Sex'], aggfunc = 'count')
```

```python
df_3.pivot_table(values='Age', index=['Pclass','Sex'], columns='SibSp', aggfunc = 'count')
```

```python
df_3.pivot_table(values='Age', index=['Pclass','Sex'], columns=['SibSp','Parch'], aggfunc = 'count')
```

```python
df_3.pivot_table(values='Fare', index='Sex', columns='Pclass', aggfunc = 'max')
```

```python
df_3.pivot_table(values='Fare', index='Sex', columns='Pclass', aggfunc = 'min')
```

```python
#This also works the same way as above syntax
pd.pivot_table(data = df_3, values='Fare', index='Sex', columns='Pclass', aggfunc = 'min')
```

### pd.melt() to create a narrow dataset from wide one

```python
df_4 = pd.read_csv('test.csv')

df_4
```

```python
df_4 = df_4[['PassengerId','Sex','Age']]

df_4
```

```python
pd.melt(df_4, id_vars='PassengerId')
```

```python
pd.melt(df_4, id_vars='PassengerId', var_name='Parameter', value_name='Data')
```

```python
pd.melt(df_4, id_vars='PassengerId', var_name='Parameter', value_name='Data').sort_values(by='PassengerId')
```
