```python
import pandas as pd
```

```python
df_1 = pd.read_csv('join_1.csv')

df_1
```

```python
df_2 = pd.read_csv('join_2.csv')

df_2
```

### pd.concat() Method to append one dataframe to another

```python
pd.concat(objs=[df_1, df_2])
```

```python
#Drop existing index and generate a new one

pd.concat(objs=[df_1, df_2], ignore_index=True)
```

```python
#Alternative to pd.concat

df_1.append(df_2)
```

```python
df_1.append(df_2, ignore_index=True)
```

```python
new_df = pd.concat(objs=[df_1, df_2], keys=['Month_1', 'Month_2'])

new_df
```

```python
new_df.loc[('Month_1',)]
```

```python
new_df.loc[('Month_1',4)]
```

```python
new_df.loc[('Month_1',4),'Name']
```

```python
new_df.loc[('Month_1',4),['Name','Salary']]
```

```python
new_df.loc[('Month_1',),'Name']
```

### Inner Joins

```python
#Inner join on single column

df_1.merge(df_2, how='inner', on='EmpID')
```

```python
df_1.merge(df_2, how='inner', on='EmpID', suffixes=('-Month_1','-Month_2'))
```

```python
#Inner join on multiple columns

df_1.merge(df_2, how='inner', on=['EmpID','Name'])
```

### OUTER JOINS

```python
df_1.merge(df_2, how='outer', on='EmpID', suffixes=('_1','_2'))
```

```python
df_1.merge(df_2, how='outer', on='EmpID', suffixes=('_1','_2'), indicator=True)
```

```python
merged=df_1.merge(df_2, how='outer', on='EmpID', suffixes=('_1','_2'), indicator=True)

merged['_merge'].value_counts()
```

```python
merged[merged['_merge']=='both']
```

```python
merged['_merge'].isin(['left_only'])
```

```python
merged[merged['_merge'].isin(['left_only'])]
```

### LEFT JOIN

```python
df_1.merge(df_2, how='left', on='EmpID')
```

```python
#It will sort based on matching column

df_1.merge(df_2, how='left', on='Name')
```

```python
df_1.merge(df_2, how='left', on='Name', sort=True)
```

### left_on and right_on parameters

```python
#This will help in joining dataframes with different column names

df_3 = df_2.copy()

df_3.columns = ['Employee_Id', 'Emp_Name', 'Emp_Salary']

df_3
```

```python
df_1.merge(df_3, how='left', left_on='EmpID', right_on='Employee_Id')
```

```python
df_1.merge(df_3, how='left', left_on='EmpID', right_on='Employee_Id').drop('Employee_Id', axis=1)
```

### Merging by Indexes using left_index and right_index paramters

```python
df_3.set_index('Employee_Id', inplace=True)

df_3
```

```python
df_1.merge(df_3, how='left', left_on='EmpID', right_index=True)
```

```python
df_4 = df_1.copy()

df_4.set_index("EmpID", inplace=True)

df_4
```

```python
df_4.merge(df_3, how='left', left_index=True, right_index=True)
```

### .join() Method

```python
#Used to concatenate dataframe columnwise

df_1.join(df_2,lsuffix='_left')
```

```python
df_1.join(df_2, how='left', lsuffix='_left', rsuffix='_right')
```

### pd.merge() Method

```python
#This is same as merge method shown in above examples but with different syntax

pd.merge(df_1, df_2, how='outer', on='EmpID')
```
