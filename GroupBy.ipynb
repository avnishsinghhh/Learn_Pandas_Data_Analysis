### .agg() Method

```python
gender = df.groupby('Sex')

gender
```

```python
gender.size()
```

```python
gender.agg({'Age':'sum', 'Fare':'mean', 'Parch':'count'})
```

```python
gender.agg(['size','sum','mean'])
```

```python
gender.agg({'Age':['sum','mean'], 'Fare':'mean', 'Parch':'count'})
```

### Iterating through the groups in GroupBy object

```python
pclass = df.groupby('Pclass')

pclass
```

```python
pclass['Age'].size()
```

```python
dframe = pd.DataFrame(columns=df.columns)

dframe
```

```python
for pclas, data in pclass:
    highest = data.nlargest(1,'Age')
    dframe = dframe.append(highest)
    
dframe
```

```python
dframe_1 = pd.DataFrame(columns=df.columns)

dframe_1
```

```python
for pclas, data in pclass:
    smallest = data.nsmallest(2,'Age')
    dframe_1 = dframe.append(smallest)
    
dframe_1
```
