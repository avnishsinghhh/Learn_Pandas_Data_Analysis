### .replace()

```python
df['Name'].str.replace('Mrs.', 'Ms.')
```

```python
df['Fare'].astype('str').str.replace('.',',')
```

### Filter Dataframe Rows with String methods. .contains(), .startswith(), .endswith()

```python
df['Name'].str.contains('Mr.')
```

```python
df[df['Name'].str.contains('Miss')]
```

```python
df[df['Name'].str.contains('Mr\.')]
```

```python
df[df['Name'].str.startswith('Myles')]
```

```python
df[df['Name'].str.endswith('Francis')]
```
