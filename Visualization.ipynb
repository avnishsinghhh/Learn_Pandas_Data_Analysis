### Bar Graphs

```python
def rank_performance(price):
    if price <= 10:
        return 'Poor'
    elif price <= 50:
        return 'Satisfactory'
    else:
        return 'Steller'
```

```python
data['Close'].apply(rank_performance)
```

```python
data['Close'].apply(rank_performance).value_counts()
```

```python
data['Close'].apply(rank_performance).value_counts().plot()
```

```python
data['Close'].apply(rank_performance).value_counts().plot(kind='line')
```

```python
data['Close'].apply(rank_performance).value_counts().plot(kind='bar')
```

```python
data['Close'].apply(rank_performance).value_counts().plot(kind='barh')
```

### Pie Charts

```python
plt.style.use('dark_background')
```
